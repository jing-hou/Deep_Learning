# data preprocessing (define functions)

```python
#functions
import math
from datetime import datetime
def status (string):
    if string=="A" or string=="C":
        return '0'
    elif string=="B" or string=="D":
        return '1'
def frequency(string):
    if string=="POPLATEK MESICNE":
        return '0'
    elif string=="POPLATEK TYDNE":
        return '1'
    elif string=="POPLATEK PO OBRATU":
        return '2'
def gender (string):
    if string=="male":
        return '0'
    elif string=="female":
        return '1'
def age(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return math.floor(abs((d1 - d2).days)/365)
def days_account(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d1 - d2).days)
```

```python
#clean data and feature enginering
#account
account['frequency']=account['frequency'].apply(frequency) #frequency_type_encoding
#loan 
loan['status']=loan['status'].apply(status)# 0 or 1
#order

order_tol= order.groupby(['account_id']).agg(oreder_count=('order_id', 'sum'), amount_sum=('amount', 'count'))

order_UVER=order[order['k_symbol'] == "UVER"][['account_id','amount']]
order_clean = pd.merge(order_tol, order_UVER, how='left', on=['account_id'])
order_clean['frac']=order_clean['amount']/order_clean['amount_sum']
order_df=order_clean.rename(columns={"amount": "amount_UVER", "frac": "aUVER/asum"})
order_df.amount_UVER.fillna(0, inplace=True)
order_df['aUVER/asum'].fillna(order_df['aUVER/asum'].mean(), inplace=True) ######### what is inside
#trans
trans['type']= pd.Series('1' if i =='PRIJEM' else '0' for i in trans['type'])
trans_before_loan=pd.merge(trans, loan, how='left', on=['account_id'], suffixes=('_trans', '_loan'))
trans_before_loan=trans_before_loan[trans_before_loan['date_loan'] >= trans_before_loan['date_trans']]
trans_before_loan
trans_balance=[]
for account_id, group in trans_before_loan.groupby('account_id'):
    group.sort_values(by=['trans_id'])
    max_balance=group['balance'].max()
    avg_balance= group['balance'].mean()
    trans_count= group['trans_id'].count()
    ratio= (avg_balance/group['amount_loan']).head(1).values[0]

    balance_final=group['balance'].tail(1).values[0]
    sum_credit=group[group['type']=='1']['amount_trans'].sum()#useless

    count_credit= group[group['type']=='1'].count().tail(1).values[0]#useless
    avg_credit=sum_credit/count_credit

    
    trans_balance.append({'account_id': account_id,'trans_count':trans_count,'max_balance': max_balance,
                          'avg_balance':avg_balance,
                          'balance/loan': ratio, 'sum_credit':sum_credit,'balance_final':balance_final,
                         'count_credit':count_credit, 'avg_credit':avg_credit})

trans_balance=pd.DataFrame(trans_balance) 
#disp-client-card
disp_card=pd.merge(disp, card, how='left', on=['disp_id'], suffixes=('_disp', '_card'))
disp_c= disp_card.groupby(['account_id']).agg(num_disp=('disp_id', 'count'))
disp_card=pd.merge(disp_card, disp_c, how='left', on=['account_id'])
##keep owner in disp_type
disp_card=disp_card[disp_card['type_disp']=='OWNER']
disp_card_client=pd.merge(disp_card, client, how='left', on=['client_id'])
disp_card_client['gender']=disp_card_client['gender'].apply(gender)#factorize()
disp_card_client.drop(['disp_id', 'client_id','type_disp','type_card','card_id'], axis=1, inplace=True)

#merge
dcc_account=pd.merge(disp_card_client, account, on=['account_id'],suffixes=('', '_account'))
dcca_loan=pd.merge(dcc_account, loan, on=['account_id'],suffixes=('_account', '_loan'))
#own credit card before having loan
dcca_loan['has_card_before_loan']=(dcca_loan['issued']<dcca_loan['date_loan'])
dcca_loan['has_card_before_loan']= pd.Series('1' if i ==True else '0' for i in dcca_loan['has_card_before_loan'])
dcca_loan.has_card_before_loan.value_counts()
#client district_id different from account district_id
dcca_loan['district_diff']=(dcca_loan.district_id==dcca_loan.district_id_account)
dcca_loan['district_diff']= pd.Series('1' if i ==False else '0' for i in dcca_loan['district_diff'])
dcca_loan['district_diff'].value_counts()
dcca_loan.drop(['issued', 'loan_id'], axis=1, inplace=True)########issued_date droped
dccal_order=pd.merge(dcca_loan, order_df, on=['account_id'],suffixes=('', '_order'))
dccalo_trans=pd.merge(dccal_order, trans_balance, on=['account_id'],suffixes=('', '_trans'))
dccalo_trans['credit/trans_count']=dccalo_trans['count_credit']/dccalo_trans['trans_count']
#dccalo_trans.drop(['count_credit', 'trans_count'], axis=1, inplace=True)
from datetime import datetime
import math
#1993-1945=48
dccalo_trans['age_loan']=dccalo_trans[['date_loan','birth_date']].apply(lambda x : age(*x), axis=1)#age when issued loan
dccalo_trans['days_account']=dccalo_trans[['date_loan','date_account']].apply(lambda x : days_account(*x), axis=1)#days bt acc & loan
dccalo_trans=pd.merge( district, dccalo_trans,on =['district_id'])
data=dccalo_trans.drop(['birth_date','district_id','district_id_account','date_account','date_loan',
                        'amount_UVER','sum_credit','count_credit','A3'], axis=1)

```

```python
from talos.utils import lr_normalizer
```

```python
# define the model 
from keras.models import Sequential
from keras.layers import Dropout, Dense
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, BatchNormalization
from keras.callbacks import TensorBoard
from keras.callbacks import ModelCheckpoint
from keras.optimizers import adam

def loan_default(x_train, y_train, x_val, y_val, params):
    
    model = Sequential()     
#     model.add(Flatten(input_shape=(X_train_resampled.shape[1],)))
# Input 0 is incompatible with layer flatten_3: expected min_ndim=3, found ndim=2
# we do not need flatten here, so we need to specify the input_dim in the first layer
    model.add(Dense(params['first_neuron'], input_dim=x_train.shape[1],activation=params['activation'],
                    kernel_initializer=params['kernel_initializer']))
    model.add(Dropout(params['dropout']))
    model.add(Dense(params['first_neuron'],activation=params['activation'],
                    kernel_initializer=params['kernel_initializer']))
    model.add(Dropout(params['dropout']))
    model.add(Dense(1, activation=params['last_activation'],
                    kernel_initializer=params['kernel_initializer']))
#     model.compile(params['optimizer'],
#                   loss=params['loss'],
#                   metrics=['Recall', 'Precision'])


# Compile model
    model.compile(
        loss=params['losses'],
       optimizer=params['optimizer'](lr=lr_normalizer(params['lr'], params['optimizer'])),
        metrics=['acc']
    )
   
    out = model.fit(x_train, y_train,
                    batch_size=params['batch_size'],
                    epochs=params['epochs'],
                    verbose=0,
                    validation_data=[x_val, y_val])
    
    return out, model
```

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import regularizers
import pandas as pd
import numpy as np


X = df8.drop(['status'],axis=1)
X = pd.get_dummies(X, drop_first = True)
y = df8['status'].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

ros = RandomOverSampler(random_state = 42)

scaler = MinMaxScaler()
scaler.fit(X_train)
scaler_X_train = scaler.transform(X_train)
scaler_X_test = scaler.transform(X_test)

X_train_resampled, y_train_resampled = ros.fit_resample(scaler_X_train, y_train)

model = keras.Sequential([
    keras.layers.Dense(10, activation=tf.nn.relu, kernel_initializer = 'uniform',),
    keras.layers.Dropout(0),
    keras.layers.Dense(10, activation=tf.nn.relu, kernel_initializer = 'uniform',),
    keras.layers.Dropout(0),
    keras.layers.Dense(1, activation=tf.nn.sigmoid),
])

opt = keras.optimizers.Adam(learning_rate=0.001)

model.compile(optimizer=opt,
              loss='binary_crossentropy',
              metrics=['Recall', 'Precision'])
   
history = model.fit(X_train_resampled, y_train_resampled, epochs=40,  \
                    validation_data = (scaler_X_test, y_test), batch_size=30,\
                    callbacks=[early_stopping], class_weight=class_weight)

graph=pd.DataFrame(history.history)
graph.plot(figsize=(10,10));
plt.legend( prop={"size":10},loc='center right')

y_pred = model.predict(scaler_X_test)
```
